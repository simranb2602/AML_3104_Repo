```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
#taking excel data in a dataframe
affdf=pd.read_csv("Algerian_forest_fires_cleaned.csv")
```

```python
# Remove leading and trailing spaces from values in the 'Classes' column
affdf['Classes'] = affdf['Classes'].str.strip()
```

```python
affdf
```

```python
# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)

# Fit and transform the 'Classes' column
encoded_classes = encoder.fit_transform(affdf[['Classes']])

# Assign the encoded values to the DataFrame
affdf[['Classes_fire', 'Classes_not_fire']] = encoded_classes

# Drop the original 'Classes' column
affdf.drop(columns=['Classes'], inplace=True)
```

```python
#dropping year column
affdf=affdf.drop(columns=['year'])
affdf
```

```python
#correlation matrix calculation
corr=affdf.corr().round(2)
corr
```

```python
#remove outliers
# Select only numerical columns for outlier detection
numerical_columns = affdf.select_dtypes(include=['float64', 'int64']).columns
 
# Calculate the IQR for each numerical column
Q1 = affdf[numerical_columns].quantile(0.25)
Q3 = affdf[numerical_columns].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Define a threshold to identify outliers (e.g., 1.5 times the IQR)
threshold = 1.5
 
# Identify outliers using the IQR method
outliers = affdf[((affdf[numerical_columns] < (Q1 - threshold * IQR)) | (affdf[numerical_columns] > (Q3 + threshold * IQR))).any(axis=1)]
 
# Display outliers
print("\nOutliers:")
print(outliers)
```

```python
# Remove outliers from the dataset
cleaned_df = affdf[~((affdf[numerical_columns] < (Q1 - threshold * IQR)) | (affdf[numerical_columns] > (Q3 + threshold * IQR))).any(axis=1)]
 
# Display the cleaned dataset
print("\nCleaned dataset:")
print(cleaned_df)
```

```python
#setting target and input variables
X=affdf.drop(columns=["FWI"])
Y=affdf['FWI']
```

```python
#split dataset into train and test
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
#scalling the data
scaler=StandardScaler()
X_train_Scale=scaler.fit_transform(X_train)
X_test_Scale=scaler.transform(X_test)
```

```python
#initializing the model
mlr=LinearRegression()
mlr.fit(X_train_Scale,Y_train)
```

```python
mlr.coef_,mlr.intercept_
```

```python
#Predicting the output variable FWI
y_pred=mlr.predict(X_test_Scale)
```

```python
#Model Evaluation
mse=mean_squared_error(Y_test,y_pred)
mae=mean_absolute_error(Y_test,y_pred)
r2=r2_score(Y_test,y_pred)
print('r2:',r2)
print('mae:',mae)
print('mse:',mse)
```

```python
#creation of pickle files to deal with unknown or new data
import pickle
pickle.dump(mlr,open('sm_mlr_assign.pkl','wb')) #wb means writebyte
pickle.dump(scaler,open('sm_scaler_assign.pkl','wb'))
```

Benefits of pickle files and their purpose

The process of pickling a machine learning model, which involves converting it into a format suitable for storage on disk, brings about several advantages:

Persistence: Pickling allows for the preservation of a trained model on disk, eliminating the need to retrain it each time the program is executed.

Portability: Serialized models can be effortlessly shared across different systems or platforms, ensuring seamless deployment in diverse environments.

Efficiency: Pickled models are compact, requiring less storage space, which proves advantageous for both storage and transmission purposes.

Scalability: Serialized models can be transferred between distinct Python environments or versions without encountering compatibility issues, facilitating smooth integration.

Security: Pickled models can be encrypted or signed, providing protection against unauthorized access or tampering.

Offline Evaluation: Serialized models support offline evaluation and testing, promoting collaboration and feedback among team members. Versioning: Serialized models can be versioned and tracked using version control systems, ensuring reproducibility and enabling the rollback to previous versions if necessary.
